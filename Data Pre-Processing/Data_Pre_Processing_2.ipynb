```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('/content/credits.csv')
credits.head()
```

```python
meta = pd.read_csv('/content/movies_metadata.csv')
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce') #coerce will set any invalid value to NaN
```

```python
 #extracting 2017 data
 meta_2017 = meta.loc[meta.year == 2017,['genres','id','title','year']]
```

```python
meta_2017
```

```python
meta_2017.dtypes
```

```python
meta_2017['id'] = meta_2017['id'].astype(int)
```

```python
data = pd.merge(meta_2017, credits, on='id')
```

```python
data
```

```python
#converting 'genre', 'cast', 'crew' columns to list
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x)) #the "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.head()
```

```python
def make_genresList(genre):
    gen = []
    for i in genre:
      if i.get('name') == 'Science Fiction':
        gen.append('Sci-Fi')
      else:
        gen.append(i.get('name'))
    if gen == []:
      return np.NaN
    else:
      return (" ".join(gen))
```

```python
data['genre_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genre_list']
```

```python
data['cast'][0]
```

```python
def get_actor1(x):
    casts = []
    for i in x:
      casts.append(i.get('name'))
    if casts == []:
      return np.NaN
    else:
      return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data.head()
```

```python
#Actor 2
def get_actor2(x):
    casts = []
    for i in x:
      casts.append(i.get('name'))
    if casts == [] or len(casts) <= 1:
      return np.NaN
    else:
      return (casts[1])

data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
#Actor 3

def get_actor3(x):
    casts = []
    for i in x:
      casts.append(i.get('name'))
    if casts == [] or len(casts) <= 2:
      return np.NaN
    else:
      return (casts[2])

data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data.head()
```

### Directors

```python
def get_directors(x):
    dt = []
    for i in x:
      if i.get('job') == 'Director':
        dt.append(i.get('name'))
    if dt == []:
      return np.NaN
    else:
      return (" ".join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data = data.drop('director_names', axis=1)
data.head()
```

```python
#feature selection
movie_data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genre_list','title']]
```

```python
movie_data.head()
```

```python
movie_data.isna().sum()
```

```python
movie_data = movie_data.dropna(how='any')
```

```python
movie_data.isna().sum()
```

```python
movie_data = movie_data.rename(columns={'genre_list':'genres'})
movie_data = movie_data.rename(columns={'title':'movie_title'})
```

```python
#converting all movie titles to lowercase
movie_data['movie_title'] = movie_data['movie_title'].str.lower()
```

```python
 movie_data['col_merge'] = movie_data['actor_1_name'] + ' ' + movie_data['actor_2_name'] + ' ' + movie_data['actor_3_name'] + ' ' + movie_data['director_name'] + ' ' + movie_data['genres']
```
