```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/movie_metadata.csv')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
data.title_year.value_counts(dropna=False).sort_index().plot(kind='barh', figsize=(15,16))
plt.xlabel("Movie Count")
plt.ylabel("Movie Year")
plt.show()
```

```python
#feature selection
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data['director_name'] = data['director_name'].fillna('Unknown')
data['actor_1_name'] = data['actor_1_name'].fillna('Unknown')
data['actor_2_name'] = data['actor_2_name'].fillna('Unknown')
data['actor_3_name'] = data['actor_3_name'].fillna('Unknown')
data.isna().sum()
```

```python
data.head()
```

```python
data['genres'] = data['genres'].str.replace('|',' ')
```

```python
data.head()
```

```python
#converting movie titles to lowercase
data['movie_title'] = data['movie_title'].str.lower()
```

```python
data.head()
```

```python
data['movie_title'][0]
```

```python
#removing null terminating char (\xa0) from movie title
data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1])
```

```python
data['movie_title'][90]
```

```python
#saving the processed data
data.to_csv('data1.csv',index=False)
```
